# Handling Missing Values 

```python
# Handling Missing Values

# 1. For categorical text fields → fill with "Unknown" or most frequent value
categorical_cols = ["Org_Airport", "Dest_Airport", "CancellationCode", "TailNum"]

for col in categorical_cols:
    if col in df.columns:
        # Option 1: Fill missing values with "Unknown"
        df[col] = df[col].fillna("Unknown")
        
        # Option 2 (alternative): Fill missing values with the most frequent (mode) value
        # df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# 2. For numeric fields related to delays → replace NaN with 0 (means no delay recorded)
delay_cols = ["CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay", "LateAircraftDelay"]

for col in delay_cols:
    if col in df.columns:
        df[col] = df[col].fillna(0)
     
```

```python
# 3. For elapsed time or airtime → fill with median (less skewed than mean)
time_cols = ["ActualElapsedTime", "AirTime"]

for col in time_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())
```

# Cancellation column

```python
# Cancellation column
# If 'CancellationCode' is null, it means the flight was NOT cancelled
if "CancellationCode" in df.columns:
    df["CancellationCode"] = df["CancellationCode"].fillna("Not Cancelled")
```

```python
# Check results 
print(df[delay_cols + ["Cancelled", "CancellationCode"]].isnull().sum())
```

```python
# 4. If still missing values remain → drop those rows (safe clean-up)
df = df.dropna()    
```

# Combine 'Date' and 'DepTime' into datetime format 

```python
# Combine 'Date' and 'DepTime' into datetime format 
# First, convert the 'Date' column to datetime type (day comes first) 
# Invalid parsing will be set as NaT
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True, errors="coerce")
```

```python
# Step 2: Convert 'DepTime' (HHMM integer) to a 4-digit string with leading zeros
# If the value is missing (NaN), replace it with "0000"
df["DepTime"] = df["DepTime"].apply(lambda x: f"{int(x):04d}" if pd.notnull(x) else "0000")
```

```python
# Step 3: Extract hour and minute from 'DepTime' string
# 'DepHour' → first two characters of DepTime
# 'DepMinute' → last two characters of DepTime
df["DepHour"] = df["DepTime"].str[:2].astype(int)
df["DepMinute"] = df["DepTime"].str[2:].astype(int)
```

```python
# Step 4: Combine 'Date', 'DepHour', and 'DepMinute' into a single datetime column
# Add hours and minutes as timedelta to the Date column
df["DepDatetime"] = df["Date"] + pd.to_timedelta(df["DepHour"], unit="h") + pd.to_timedelta(df["DepMinute"], unit="m")
```

```python
# Derived Features 

# Extract the month number from departure datetime (1–12)
df["Month"] = df["DepDatetime"].dt.month

# Extract the day name of the week (Monday, Tuesday, etc.)
df["DayOfWeek"] = df["DepDatetime"].dt.day_name()

# Extract the hour of departure
df["Hour"] = df["DepDatetime"].dt.hour

# Create a route string by combining Origin and Destination (e.g., ATL-LAX)
df["Route"] = df["Origin"] + "-" + df["Dest"]
```

```python
# Check results 
# Display the first few rows of the relevant columns to verify the derived features
print(df[["Date", "DepTime", "DepDatetime", "Month", "DayOfWeek", "Hour", "Route"]].head())
```

# Convert HHMM integers to HH:MM:SS string

```python
# ---- Convert HHMM integers to HH:MM:SS string ----
def hhmm_to_hhmmss(series):
    """
    Convert HHMM integers/floats (e.g., 1323 -> 13:23:00) to HH:MM:SS string.
    Handles missing or invalid values safely.
    """
    def parse_val(x):
        try:
            x_int = int(float(x))
            # If invalid HHMM, set to 00:00
            if x_int < 0 or x_int > 2359:
                return "00:00:00"
            hhmm_str = str(x_int).zfill(4)
            return f"{hhmm_str[:2]}:{hhmm_str[2:]}:00"
        except:
            return "00:00:00"
    return series.apply(parse_val)

# ---- Apply conversion to update same columns ----
for col in ["DepTime", "ArrTime", "CRSArrTime", "ActualElapsedTime"]:
    df[col] = hhmm_to_hhmmss(df[col])

# ---- Check results ----
print(df[["DepTime", "ArrTime", "CRSArrTime", "ActualElapsedTime"]].head())
```
