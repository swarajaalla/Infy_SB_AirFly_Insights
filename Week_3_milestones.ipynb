Week 3: Univariate and Bivariate Visual Analysis

• Top airlines, routes, and busiest months 

• Flight distribution by day, time, and airport
 
• Plot bar charts, histograms, boxplots, and line plots 

```python
#STEP 1: Import and Load Final Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load final preprocessed dataset
final_path = "/Volumes/workspace/default/airlines/Flight_delay_final.csv"
df = pd.read_csv(final_path)

print("✅ Dataset loaded:", df.shape)
df.head()
```

```python
#STEP 2: Univariate Analysis (Week 3)
#1️⃣ Top Airlines by Number of Flights
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Airline', order=df['Airline'].value_counts().index, palette='crest')
plt.title("Top Airlines by Number of Flights")
plt.xlabel("Airline")
plt.ylabel("Flight Count")
plt.xticks(rotation=45)
plt.show()

```

```python
#2️⃣ Top 10 Busiest Routes
top_routes = df['Route'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_routes.values, y=top_routes.index, palette='viridis')
plt.title("Top 10 Busiest Routes")
plt.xlabel("Number of Flights")
plt.ylabel("Route")
plt.show()

```

```python
#3️⃣ Flights by Month
month_counts = df['Month'].value_counts().sort_index()
plt.figure(figsize=(8,5))
plt.plot(month_counts.index, month_counts.values, marker='o', color='steelblue')
plt.title("Flights Distribution by Month")
plt.xlabel("Month")
plt.ylabel("Number of Flights")
plt.grid(alpha=0.3)
plt.show()


```

```python
#4️⃣ Flight Distribution by Day of week
import matplotlib.pyplot as plt
import seaborn as sns

# Create a mapping for day numbers to names
day_labels = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Count flights by day
day_counts = df['DayOfWeek'].value_counts().sort_index()

# Plot
plt.figure(figsize=(9,5))
sns.barplot(x=day_counts.index, y=day_counts.values, palette='crest')

# Replace 0–6 tick labels with names
plt.xticks(ticks=range(7), labels=day_labels, rotation=45)
plt.title("Flights by Day of Week", fontsize=14)
plt.xlabel("Day of Week")
plt.ylabel("Number of Flights")

# Annotate counts on bars
ax = plt.gca()
for p in ax.patches:
    ax.annotate(int(p.get_height()), 
                (p.get_x() + p.get_width()/2, p.get_height()), 
                ha='center', va='bottom', fontsize=9, color='black')

plt.show()

# Optional: print counts for verification
print(day_counts)





```

```python
#5️⃣Flights by Hour of Departure
hourly_counts = df['Hour'].value_counts().sort_index()
plt.figure(figsize=(10,5))
plt.plot(hourly_counts.index, hourly_counts.values, marker='o', color='teal')
plt.title("Flight Frequency by Departure Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Flights")
plt.grid(alpha=0.3)
plt.show()
```

```python
#6️⃣ Flights by Origin Airport
# Top 10 busiest origin airports by name
top_org_airports = df['Org_Airport'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_org_airports.values, y=top_org_airports.index, palette='flare')

plt.title("Top 10 Busiest Origin Airports", fontsize=14)
plt.xlabel("Number of Flights", fontsize=12)
plt.ylabel("Origin Airport", fontsize=12)

# Annotate flight counts
for index, value in enumerate(top_org_airports.values):
    plt.text(value + 200, index, str(value), va='center')

plt.tight_layout()
plt.show()

```

```python
#7️⃣ Average Delay by Airline (Bivariate)
# Calculate mean delay per airline
avg_delay = df.groupby('Airline')['ArrDelay'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=avg_delay.values, y=avg_delay.index, palette='coolwarm')
plt.title("Average Arrival Delay by Airline", fontsize=14)
plt.xlabel("Average Arrival Delay (minutes)", fontsize=12)
plt.ylabel("Airline", fontsize=12)

# Annotate bars
for i, val in enumerate(avg_delay.values):
    plt.text(val + 1, i, f"{val:.1f}", va='center')

plt.tight_layout()
plt.show()

```

```python
#8️⃣ Correlation Heatmap (Continuous Features)
plt.figure(figsize=(6,4))
sns.heatmap(df[['DepDelay','ArrDelay','AirTime','CRSElapsedTime']].corr(), annot=True, cmap='YlGnBu')
plt.title("Correlation Heatmap")
plt.show()
```
