```python

import matplotlib.pyplot as plt
# sample data
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(y1,x,color='blue')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color='red')
plt.title("Plot 1")

plt.subplot(2,2,4)
plt.plot(y2,x,color='Black')
plt.title("Plot 1")
```
