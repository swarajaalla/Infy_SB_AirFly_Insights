```python
#formatting the Date datatype
print(df["Date"].dtype)
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y", errors="coerce")

print(df["Date"].dtype)
df.head()
```

```python
#Convert DepTime (HHMM int) into string with leading zeros
df["DepTime"] = df["DepTime"].apply(lambda x: f"{int(x):04d}" if pd.notnull(x) else "0000")
```

```python
print("\nMemory Usage before Optimization:")
total_mem_mb = df.memory_usage(deep=True).sum() / (1024 * 1024)
print(f"Total memory usage: {total_mem_mb:.2f} MB")
```

```python
# Optimize memory usage by converting columns to appropriate dtypes
for col in df.select_dtypes(include=['int64']).columns:
    df[col] = pd.to_numeric(df[col], downcast='integer')

for col in df.select_dtypes(include=['float64']).columns:
    df[col] = pd.to_numeric(df[col], downcast='float')

for col in df.select_dtypes(include=['object']).columns:
    num_unique_values = df[col].nunique()
    num_total_values = len(df[col])
    if num_unique_values / num_total_values < 0.5:
        df[col] = df[col].astype('category')

print("\nMemory Usage after Optimization:")
total_mem_mb = df.memory_usage(deep=True).sum() / (1024 * 1024)
print(f"Total memory usage: {total_mem_mb:.2f} MB")
```

```python
#check for duplicates and removing them from the dataset
print(df.duplicated().sum())
df_new=df.drop_duplicates()
print(df_new.duplicated().sum())
```

```python
df_new.isnull().sum()
```

```python
df_new = df_new.copy()  # <-- now df_new is a standalone DataFrame
```

```python
# Fill numeric columns with median and mean, categorical with mode
for col in df_new.columns:
    if pd.api.types.is_numeric_dtype(df_new[col]):
        median_val = df_new[col].median()
        mean_val = df_new[col].mean()
        df_new.loc[:, col] = df_new[col].fillna(median_val).fillna(mean_val)
    elif pd.api.types.is_categorical_dtype(df_new[col]) or df_new[col].dtype == object:
        mode_val = df_new[col].mode().iloc[0] if not df_new[col].mode().empty else None
        if mode_val is not None:
            df_new.loc[:, col] = df_new[col].fillna(mode_val)
```

```python
df_new.isnull().sum()
```

```python
# changing the columns DepTime, ArrTime, CRSArrTime to correct format of time (e.g., 1832 to 18:32)
for col in ["DepTime", "ArrTime", "CRSArrTime"]:
    if col in df_new.columns:
        df_new[col] = df_new[col].apply(lambda x: f"{int(x):04d}" if pd.notnull(x) and str(x).isdigit() else "0000")
        df_new[col] = df_new[col].apply(lambda x: f"{x[:2]}:{x[2:]}" if len(x) == 4 else x)

# changing the columns ActualElapsedTime, CRSElapsedTime, AirTime from mins to hh:mm
def mins_to_hhmm(x):
    if pd.isnull(x):
        return None
    x = int(x)
    hours = x // 60
    minutes = x % 60
    return f"{hours:02d}:{minutes:02d}"

for col in ["ActualElapsedTime", "CRSElapsedTime", "AirTime"]:
    if col in df_new.columns:
        df_new[col] = df_new[col].apply(mins_to_hhmm)
```

```python
#add columns which are useful for further analysis
df_new["Month"] = df_new["Date"].dt.month
df_new["DayOfWeek"] = df_new["Date"].dt.day_name()
df_new["Route"] = df_new["Origin"].astype(str) + "-" + df_new["Dest"].astype(str)
```
