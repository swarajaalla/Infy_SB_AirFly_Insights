```python
# Importing required libraries
import pandas as pd                 # For handling datasets
import matplotlib.pyplot as plt     # For plotting charts
import seaborn as sns               # For advanced and good-looking visualizations

# Get Top 10 Airlines
# 'value_counts()' counts how many times each airline appears in the 'Airline' column
# 'head(10)' selects the top 10 airlines with the highest flight counts
top_airlines = df['Airline'].value_counts().head(10)

# Create a figure of size 10 (width) x 6 (height)
plt.figure(figsize=(10,6))

# Draw a horizontal bar plot
# x = number of flights for each airline
# y = airline names
# palette = color scheme for bars
sns.barplot(x=top_airlines.values, y=top_airlines.index, palette="viridis")

# Add title to the chart
plt.title("Top 10 Airlines by Number of Flights")

# Label the x-axis
plt.xlabel("Number of Flights")

# Label the y-axis
plt.ylabel("Airline")

# Display the final plot
plt.show()
```

```python
# Top Routes
top_routes = df['Route'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_routes.values, y=top_routes.index, palette="plasma")
plt.title("Top 10 Busiest Routes")
plt.xlabel("Number of Flights")
plt.ylabel("Route (Origin  Destination)")
plt.show()
```

```python
# Busiest Months
month_counts = df['Month'].value_counts().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=month_counts.index, y=month_counts.values, marker="o", linewidth=2.5)
plt.title("Number of Flights per Month")
plt.xlabel("Month (1 = Jan ... 12 = Dec)")
plt.ylabel("Number of Flights")
plt.xticks(range(1,13))
plt.show()
```

```python
# Column used: DayOfWeek
#Purpose: See which days have the most flights (e.g., Fridays may be busier).
# count flights by day of week
flights_by_day = df['DayOfWeek'].value_counts().sort_index()
plt.figure(figsize=(8,5))
sns.barplot(x=flights_by_day.index, y=flights_by_day.values, palette="coolwarm")
plt.title("Flight Distribution by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Flights")
plt.show()
```

```python
# Columns used: DepHour (or Hour)
#Purpose: See when flights mostly depart (e.g., morning vs evening peaks).
# Count flights by Departure Hour

flights_by_hour = df['DepHour'].value_counts().sort_index()
plt.figure(figsize=(10,5))
sns.lineplot(x=flights_by_hour.index, y=flights_by_hour.values, marker="o")
plt.title("Flight Distribution by Hour of Day")
plt.xlabel("Departure Hour (0 = Midnight ... 23 = 11PM)")
plt.ylabel("Number of Flights")
plt.xticks(range(0,24))
plt.show()
```

```python
# Top 10 Origin Airports
top_origin = df['Origin'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_origin.values, y=top_origin.index, palette="magma")
plt.title("Top 10 Origin Airports by Departures")
plt.xlabel("Number of Flights")
plt.ylabel("Origin Airport")
plt.show()

# Top 10 Destination Airports
top_dest = df['Dest'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_dest.values, y=top_dest.index, palette="cividis")
plt.title("Top 10 Destination Airports by Arrivals")
plt.xlabel("Number of Flights")
plt.ylabel("Destination Airport")
plt.show()
```

```python
# Pivot table: Rows = DayOfWeek, Columns = DepHour, Values = Count of Flights
flights_day_hour = df.pivot_table(index='DayOfWeek', 
                                  columns='DepHour', 
                                  values='FlightNum', 
                                  aggfunc='count').fillna(0)

plt.figure(figsize=(15,6))
sns.heatmap(flights_day_hour, cmap="YlGnBu", annot=False, cbar=True)
plt.title("Flight Distribution by Day of Week and Hour of Day")
plt.xlabel("Departure Hour (0-23)")
plt.ylabel("Day of Week")
plt.show()
```

```python
# Pivot table: Rows = Origin Airport, Columns = DayOfWeek
flights_airport_day = df.pivot_table(index='Origin', 
                                     columns='DayOfWeek', 
                                     values='FlightNum', 
                                     aggfunc='count').fillna(0)

# Show only top 10 busiest airports
top_airports = df['Origin'].value_counts().head(10).index
flights_airport_day = flights_airport_day.loc[top_airports]

plt.figure(figsize=(12,6))
sns.heatmap(flights_airport_day, cmap="mako", annot=True, fmt="g")
plt.title("Flights by Origin Airport and Day of Week (Top 10 Airports)")
plt.xlabel("Day of Week")
plt.ylabel("Origin Airport")
plt.show()
```

```python
# Pivot table: Rows = Origin Airport, Columns = DepHour
flights_airport_hour = df.pivot_table(index='Origin', 
                                      columns='DepHour', 
                                      values='FlightNum', 
                                      aggfunc='count').fillna(0)

# Show only top 10 busiest airports
flights_airport_hour = flights_airport_hour.loc[top_airports]

plt.figure(figsize=(14,6))
sns.heatmap(flights_airport_hour, cmap="rocket", annot=False, cbar=True)
plt.title("Flights by Origin Airport and Hour of Day (Top 10 Airports)")
plt.xlabel("Departure Hour")
plt.ylabel("Origin Airport")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("whitegrid")

# Create a figure with 2 rows and 2 columns
fig, axs = plt.subplots(2, 2, figsize=(18,12))

# ----------- 1. Bar Chart: Top 10 Airlines -----------
top_airlines = df['Airline'].value_counts().head(10)
sns.barplot(x=top_airlines.values, y=top_airlines.index, palette="viridis", ax=axs[0,0])
axs[0,0].set_title("Top 10 Airlines by Number of Flights", fontsize=14)
axs[0,0].set_xlabel("Number of Flights")
axs[0,0].set_ylabel("Airline")

# ----------- 2. Histogram: Arrival Delays -----------
sns.histplot(df['ArrDelay'], bins=50, kde=True, color="skyblue", ax=axs[0,1])
axs[0,1].set_title("Distribution of Arrival Delays", fontsize=14)
axs[0,1].set_xlabel("Arrival Delay (minutes)")
axs[0,1].set_ylabel("Number of Flights")
axs[0,1].set_xlim(-50, 300)  # focus on main delay range

# ----------- 3. Boxplot: Arrival Delays by Airline -----------
sns.boxplot(x='Airline', y='ArrDelay', data=df, ax=axs[1,0])
axs[1,0].set_title("Arrival Delays by Airline", fontsize=14)
axs[1,0].set_xlabel("Airline")
axs[1,0].set_ylabel("Arrival Delay (minutes)")
axs[1,0].set_xticklabels(axs[1,0].get_xticklabels(), rotation=45)
axs[1,0].set_ylim(-50, 300)

# ----------- 4. Line Plot: Flights per Month -----------
flights_by_month = df['Month'].value_counts().sort_index()
sns.lineplot(x=flights_by_month.index, y=flights_by_month.values, marker="o", ax=axs[1,1])
axs[1,1].set_title("Number of Flights per Month", fontsize=14)
axs[1,1].set_xlabel("Month (1=Jan ... 12=Dec)")
axs[1,1].set_ylabel("Number of Flights")
axs[1,1].set_xticks(range(1,13))

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

CarrierDelay  Delay caused by the airline (crew, maintenance)

WeatherDelay  Delay caused by weather conditions

NASDelay  National Air System delays (air traffic control, system issues)

SecurityDelay  Delay caused by security issues

LateAircraftDelay  Delay caused by the previous flight of the same aircraft

```python

# Select relevant columns
delay_columns = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
airline_delay = df.groupby('Airline')[delay_columns].mean().reset_index()

# Melt dataframe for easier plotting
airline_delay_melted = airline_delay.melt(id_vars='Airline', 
                                          value_vars=delay_columns, 
                                          var_name='DelayType', 
                                          value_name='AverageDelay')

plt.figure(figsize=(12,6))
sns.barplot(x='Airline', y='AverageDelay', hue='DelayType', data=airline_delay_melted)
plt.title("Average Delay Causes by Airline")
plt.ylabel("Average Delay (minutes)")
plt.xlabel("Airline")
plt.xticks(rotation=45)
plt.legend(title="Delay Type")
plt.show()
```

```python
# Summary stats for selected delays
delay_cols = ['CarrierDelay', 'WeatherDelay', 'NASDelay']
df[delay_cols].describe()
```

```python
#Distribution Plots like carrier delay,weather delay,NAS delay (Histograms)
plt.figure(figsize=(16,5))
for i, col in enumerate(delay_cols):
    plt.subplot(1, 3, i+1)
    sns.histplot(df[col], bins=50, kde=True, color='skyblue')
    plt.title(f"Distribution of {col}")
    plt.xlabel("Delay (minutes)")
    plt.xlim(0, 200)  # focus on main range
    plt.ylabel("Number of Flights")

plt.tight_layout()
plt.show()
```

```python
# Group by Airline
avg_delay_airline = df.groupby('Airline')[delay_cols].mean().reset_index()

# Melt for plotting
avg_delay_melted = avg_delay_airline.melt(id_vars='Airline', 
                                          value_vars=delay_cols, 
                                          var_name='DelayType', 
                                          value_name='AverageDelay')

plt.figure(figsize=(12,6))
sns.barplot(x='Airline', y='AverageDelay', hue='DelayType', data=avg_delay_melted)
plt.title("Average Carrier, Weather, and NAS Delays by Airline")
plt.ylabel("Average Delay (minutes)")
plt.xlabel("Airline")
plt.xticks(rotation=45)
plt.legend(title="Delay Type")
plt.show()
```

```python
#Average Delay by Hour of Day

# Group by departure hour and compute average arrival delay
avg_delay_hour = df.groupby('DepHour')['ArrDelay'].mean().reset_index()

plt.figure(figsize=(12,5))
sns.lineplot(x='DepHour', y='ArrDelay', data=avg_delay_hour, marker='o', color='red')
plt.title("Average Arrival Delay by Hour of Day")
plt.xlabel("Departure Hour (0-23)")
plt.ylabel("Average Arrival Delay (minutes)")
plt.xticks(range(0,24))
plt.grid(True)
plt.show()
```

```python
#Boxplot of Delays by Hour
plt.figure(figsize=(15,6))
sns.boxplot(x='DepHour', y='ArrDelay', data=df)
plt.title("Arrival Delay Distribution by Hour of Day")
plt.xlabel("Departure Hour")
plt.ylabel("Arrival Delay (minutes)")
plt.ylim(-50, 300)  # focus on main delay range
plt.show()
```

```python
# Compute average arrival delay per airport
#Average Delay by Origin Airport
avg_delay_airport = df.groupby('Origin')['ArrDelay'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=avg_delay_airport.values, y=avg_delay_airport.index, palette="magma")
plt.title("Top 10 Airports by Average Arrival Delay")
plt.xlabel("Average Arrival Delay (minutes)")
plt.ylabel("Origin Airport")
plt.show()
```

```python
# Heatmap: Delays by Hour & Day of Week
# Pivot table: Rows = DayOfWeek, Columns = DepHour, Values = Avg ArrDelay
delay_day_hour = df.pivot_table(index='DayOfWeek', columns='DepHour', values='ArrDelay', aggfunc='mean')

plt.figure(figsize=(15,6))
sns.heatmap(delay_day_hour, cmap="Reds", annot=False)
plt.title("Average Arrival Delay by Day of Week and Hour of Day")
plt.xlabel("Departure Hour")
plt.ylabel("Day of Week")
plt.show()
```
