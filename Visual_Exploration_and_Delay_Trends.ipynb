```python
# 2. Count flights per airline
airline_counts = df['Airline'].value_counts().reset_index()
airline_counts.columns = ['Airline', 'Number_of_Flights']
```

```python
# 3. Display top 10 airlines
top_airlines = airline_counts.head(10)
print(top_airlines)
```

```python
# 4. Plot using matplotlib
plt.figure(figsize=(12,6))
plt.bar(top_airlines['Airline'], top_airlines['Number_of_Flights'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Airlines by Number of Flights")
plt.xlabel("Airline")
plt.ylabel("Number of Flights")
plt.tight_layout()
plt.show()
```

```python
# 5. Plot using seaborn
plt.figure(figsize=(12,6))
sns.barplot(x='Number_of_Flights', y='Airline', data=top_airlines, palette='viridis')
plt.title("Top 10 Airlines by Number of Flights")
plt.xlabel("Number of Flights")
plt.ylabel("Airline")
plt.show()
```

```python
# Count flights by day of the week and sort by weekday order
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
flights_by_day = df['DayOfWeek'].value_counts().reindex(weekday_order)

# Print the counts
print("\nFlights by Day of Week")
print(flights_by_day)
```

```python
#Top 10 longest flights (by Distance)
longest_flights = df.nlargest(10, 'Distance')[['Airline', 'FlightNum', 'Origin', 'Dest', 'Distance']]
print("\nTop 10 Longest Flight")
print(longest_flights)
```

```python
# Take top 10 long flights
top_long_flights = longest_flights[['FlightNum', 'Origin', 'Dest', 'Distance']].head(10)

# Create a label for each flight as "Origin → Dest"
top_long_flights['Route'] = top_long_flights['Origin'] + " → " + top_long_flights['Dest']
print(top_long_flights)
```

```python
# Plot using matplotlib
plt.figure(figsize=(12,6))
bars = plt.bar(top_long_flights['Route'], top_long_flights['Distance'], color='orange')
plt.title("Top 10 Long Flights (>1000 miles)")
plt.xlabel("Route")
plt.ylabel("Distance (miles)")
plt.xticks(rotation=45)
```

```python
# Group by departure hour and calculate mean delay
avg_delay = df.groupby('DepHour')['ArrDelay'].mean().reset_index()

# Plot aggregated data
sns.lineplot(x='DepHour', y='ArrDelay', data=avg_delay, marker='o', color='tomato')

plt.title("Average Arrival Delay vs Departure Hour")
plt.xlabel("Departure Hour")
plt.ylabel("Average Arrival Delay (minutes)")
plt.grid(True)
plt.show()
```

```python
# Step 1: Select the relevant delay columns
delay_cols = ['WeatherDelay', 'NASDelay', 'LateAircraftDelay']

# Step 2: Calculate total delay time (sum of each delay type)
df_delay = df[delay_cols].sum().sort_values(ascending=False)

# Step 3: Plot using Matplotlib bar chart
plt.figure(figsize=(10,6))  # set figure size

# Create bar chart
plt.bar(df_delay.index, df_delay.values, color=['skyblue', 'orange', 'lightgreen'])

# Step 4: Add title and labels
plt.title("Total Delay Time by Type", fontsize=14)
plt.xlabel("Delay Type", fontsize=12)
plt.ylabel("Total Minutes of Delay", fontsize=12)
```

```python
# Group the data by 'Route' and calculate the mean arrival delay for each route
# Then sort in descending order and select the top 10 routes with the highest average delay
route_delay = df.groupby('Route')['ArrDelay'].mean().sort_values(ascending=False).head(10)

# Set the figure size for better readability
plt.figure(figsize=(12,6))

# Create a bar chart for the top 10 routes with the highest average delay
plt.bar(route_delay.index, route_delay.values, color='coral')

# Add title and axis labels
plt.title("Top 10 Routes with Highest Average Delay")
plt.xlabel("Route")
plt.ylabel("Average Delay (minutes)")

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Select the columns related to different types of delays
delay_cols = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']

# Calculate the total delay minutes for each delay type across all flights
delay_totals = df[delay_cols].sum()

# Create a pie chart to visualize the proportion of each delay type
plt.pie(
    delay_totals,                         # Values to plot
    labels=delay_totals.index,            # Labels (delay types)
    autopct='%1.1f%%',                    # Display percentage on the chart
    startangle=140,                       # Rotate start of pie for better readability
    colors=['gold', 'skyblue', 'lightgreen', 'pink', 'orange']  # Custom colors for each slice
)

# Add a descriptive title
plt.title("Proportion of Delay Time by Type")

# Display the plot
plt.show()
```

```python
# Select relevant columns
delay_cols = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay', 'Airline']

# Group by Airline and calculate mean delays
airline_delays = df.groupby('Airline')[delay_cols[:-1]].mean().reset_index()

# Visualize
plt.figure(figsize=(12,6))
airline_delays.set_index('Airline').plot(kind='bar', stacked=True, figsize=(14,7))
plt.title("Average Delay by Cause per Airline")
plt.ylabel("Minutes of Delay")
plt.xticks(rotation=45)
plt.show()

```

```python
# Average delay by departure hour
hourly_delays = df.groupby('DepHour')[['CarrierDelay', 'WeatherDelay', 'NASDelay']].mean().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=hourly_delays, x='DepHour', y='CarrierDelay', label='Carrier')
sns.lineplot(data=hourly_delays, x='DepHour', y='WeatherDelay', label='Weather')
sns.lineplot(data=hourly_delays, x='DepHour', y='NASDelay', label='NAS')
plt.title("Average Delay by Time of Day")
plt.xlabel("Departure Hour")
plt.ylabel("Average Delay (minutes)")
plt.legend()
plt.show()
```

```python
# Average delay per airport
airport_delays = df.groupby('Origin')[['CarrierDelay','WeatherDelay','NASDelay']].mean().sort_values(by='CarrierDelay', ascending=False).reset_index()

plt.figure(figsize=(12,6))
airport_delays.set_index('Origin')[['CarrierDelay','WeatherDelay','NASDelay']].plot(kind='bar', stacked=True, figsize=(14,7))
plt.title("Average Delay by Airport")
plt.ylabel("Minutes of Delay")
plt.xticks(rotation=45)
plt.show()
```
