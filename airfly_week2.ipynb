```python
df['Date']=pd.to_datetime(df['Date'],format="mixed")
df['month']=df['Date'].dt.month
df['Day']=df.Date.dt.dayofweek+1
```

```python
df.columns.to_list
```

```python
df['Total_delay']=df['ArrDelay'].fillna(0)+df['DepDelay'].fillna(0)
```

```python
display(df)
```

```python
df.DepTime=pd.to_datetime(df.DepTime.astype(str).str.zfill(4),format="%H%M",errors='coerce').dt.strftime('%H:%M')
df.ArrTime=pd.to_datetime(df.ArrTime.astype(str).str.zfill(4),format="%H%M",errors='coerce').dt.strftime('%H:%M')
df.CRSArrTime=pd.to_datetime(df.CRSArrTime.astype(str).str.zfill(4),format="%H%M",errors='coerce').dt.strftime('%H:%M')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.DepTime=df.DepTime.bfill()
```

```python
df.ArrTime=df.ArrTime.ffill()
```

```python
df.Org_Airport=df.Org_Airport.fillna('Unknown')
df.Dest_Airport=df.Dest_Airport.fillna('Unknown')
```

```python
df.to_csv('Air_lines_cleaned.csv',index=False)
```
